```python
# import opencv library
import cv2
import matplotlib.pyplot as plt
from skimage.measure import compare_ssim as ssim
import numpy as np
```

```python
capture = cv2.VideoCapture("Footage/magnesis_btb0_loaded_trimmed.mp4")
capture.set(3, 640)
capture.set(4, 480)
success, img = capture.read()
print(success)
f_num = 0
frames = []

total_frames = int(capture.get(cv2.CAP_PROP_FRAME_COUNT))
sample_rate = total_frames

for f_num in range(0, total_frames, 1):
    print(f_num)
    capture.set(cv2.CAP_PROP_POS_FRAMES, f_num)
    _, image = capture.read()
    resize = cv2.resize(image, (640, 480))
    frames.append(resize)
```

```python
np.save("magnesis_btb0_loaded_array_trimmed", frames)
```

```python
sum_load = 0
loads = []
for i in range(1, len(frames)):
    if (ssim(frames[i-1], frames[i], multichannel = True) >= 0.95):
        sum_load += 1
    elif (sum_load != 0):
        loads.append(sum_load)
        sum_load = 0
    else:
        sum_load = 0
    print("the ssim of " + str(i) + " and " + str(i-1) + " is :" + str(ssim(frames[i -1], frames[i], multichannel = True)))
```
