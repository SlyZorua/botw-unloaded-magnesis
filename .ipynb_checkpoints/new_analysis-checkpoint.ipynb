```python
# import opencv library
import cv2
import matplotlib.pyplot as plt
from skimage.measure import compare_ssim as ssim
import numpy as np
```

```python
capture = cv2.VideoCapture("Footage/magnesis_btb0_loaded.mkv")
capture.set(3, 640)
capture.set(4, 480)
success, img = capture.read()
f_num = 0
frames = []

total_frames = int(capture.get(cv2.CAP_PROP_FRAME_COUNT))
sample_rate = total_frames

for f_num in range(0, total_frames, 1):
    print(f_num)
    capture.set(cv2.CAP_PROP_POS_FRAMES, f_num)
    _, image = capture.read()
    resize = cv2.resize(image, (640, 480))
    frames.append(resize)
```

```python
plt.imshow(cv2.cvtColor(frames[2000], cv2.COLOR_BGR2RGB))
```

```python
np.save("magnesis_btb0_loaded_array", frames)
```

```python
for i in range(1, len(frames)):
    print("the ssim of " + str(i) + " and " + str(i-1) + " is :" + str(ssim(frames[i -1], frames[i], multichannel = True)))
```

```python
plt.imshow(cv2.cvtColor(frames[1636], cv2.COLOR_BGR2RGB))
plt.imshow(cv2.cvtColor(frames[1635], cv2.COLOR_BGR2RGB))
```

```python
fig = plt.figure("Images")
plt.suptitle("SSIM of Frame 1635 and 1636: %.10f" % ssim(frames[1635], frames[1636], multichannel = True))
# show first image
ax = fig.add_subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(frames[1635], cv2.COLOR_BGR2RGB))
plt.axis("off")
# show the second image
ax = fig.add_subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(frames[1636], cv2.COLOR_BGR2RGB))
plt.axis("off")
# show the images
plt.show()
```
